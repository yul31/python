#### 문자 데이터를 숫자로 변환

### 레이블 인코딩
- 숫자의 크고 작음에 대한 특성이 작용함
- 회귀와 같이 연속된 실수를 다루는 알고리즘에서는 1,2,3,... 커지는 숫자가 숫자에 따른 순서나 중요도로 인식될 수 있어서 잘못된 결과가 나올 수 있음

```python
x['cap-shape'].unique() #feature(column) 확인
```

```python
x['cap-shape'].value_counts() #각 feature별 개수 확인
```

```python
x['cap-shape']
```

```python
x['cap-shape'].map({'x':0, 'f':1, 'k':2, 'b':3, 's':4, 'c':5})
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
encoder.fit(x['cap-shape']) #레이블인코딩은 컬럼별로 다 따로 인코딩 학습을 해야한다
labels = encoder.transform(x['cap-shape']) #인코더의 unique값들을 숫자로 변환
```

```python
print(labels)
```

### 원핫인코딩
- 모든 데이터를 0과 1로 변환
- 기계는 모든 데이터를 이진법으로 처리하기 때문에 레이블인코딩 보다는 원핫인코딩을 사용함

```python
pd.set_option('display.max_columns', None) #column 다 보이게 출력
#None 자리에 숫자 넣으면 숫자개수만큼 보이게 출력
```

```python
x
```

```python
x_one_hot = pd.get_dummies(x)
```

```python
x_one_hot
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_one_hot, y, test_size=0.3, random_state=0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
tr = DecisionTreeClassifier()
tr.fit(x_train, y_train)
tr.score(x_test, y_test) #정확도 100% -> 아래의 하이퍼파라미터 조정으로 과대적합임을 알 수 있음
```

### 하이퍼파라미터 조정

```python
tr2 = DecisionTreeClassifier(max_depth=2, max_leaf_nodes=3, min_samples_leaf=4)
```

```python
tr2.fit(x_train, y_train)
tr2.score(x_test, y_test) #정확도 94% 
```

```python
train_list = []
test_list = []

for k in range(1, 11, 1) : #1~11까지, 1씩 증가
    
    #모델 생성 및 하이퍼파라미터에 변수 넣음
    tr3 = DecisionTreeClassifier(max_depth = k) #k의 증가에 따른 정확도 변화 예측
    tr3.fit(x_train, y_train)
    
    train_score = tr3.score(x_train, y_train)
    train_list.append(train_score)
    
    test_score = tr3.score(x_test, y_test)
    test_list.append(test_score)
```

```python
plt.figure(figsize=(15,5))
plt.plot(range(1,11,1), train_list, label='train')
plt.plot(range(1,11,1), test_list, label='test')
plt.legend()
plt.show()
#max_depth: 6부터 과대적합 -> 6 미만의 수로 설정해야 함
```

### 특성선택
- 지도학습 모델의 특성중요도(각 특성들은 0~1 사이의 중요도 값을 가짐, 모든 중요도의 합은 1이 됨)
- 0이 나오는 컬럼은 모델이 정답을 예측하는데 전혀 도움이 되지 않는 특성
- 1은 완벽하게 정답을 예측해내는 특성 (1이 나오면 나머지 특성은 필요가 없다)
